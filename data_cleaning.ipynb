```python
#import libraries
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
#importing dataset

ds=pd.read_csv('college_dropout_risk_rich_uncleaned.csv')
ds
```

```python
ds.columns
```

```python
#handling current_gpa column

ds['current_gpa'].unique()
```

```python
ds['current_gpa'] = pd.to_numeric(ds['current_gpa'], errors='coerce') 

x=ds['current_gpa'].mean()
print(x)
ds['current_gpa'].fillna(x,inplace=True)
```

```python
ds['attendance_rate'] = pd.to_numeric(ds['attendance_rate'], errors='coerce')

x=ds['attendance_rate'].mean()
print(x)
ds['attendance_rate'].fillna(x,inplace=True)
```

```python
#handling 'study_hours_per_week' column

ds['study_hours_per_week'].unique()
```

```python
ds['study_hours_per_week'] = pd.to_numeric(ds['study_hours_per_week'], errors='coerce')

x=ds['study_hours_per_week'].median()
print(x)
ds['study_hours_per_week'].fillna(x,inplace=True)
```

```python
ds.columns
```

```python
#handling 'financial_stress' column

ds['financial_stress'].unique()
```

```python
ds['financial_stress'] = pd.to_numeric(ds['financial_stress'], errors='coerce')
```

```python
ds.head(20)
```

```python
#handling 'mental_halth._score' column

ds['mental_health_score'].unique()
```

```python
ds['mental_health_score'] = pd.to_numeric(ds['mental_health_score'], errors='coerce')
```

```python
#handling part_time_job column

ds['part_time_job'].unique()
```

```python
ds['part_time_job'] = ds['part_time_job'].replace({'Yes': 1, 'No': 0})
```

```python
#handling 'parental_support' column
ds['parental_support'].unique()
```

```python
ds['parental_support'] = ds['parental_support'].replace({' Medium': 'medium', 'LOW': 'low', 'High': 'high'})
```

```python
#handling dropout_risk column

ds['dropout_risk'].unique()
```

```python
ds['dropout_risk'] = pd.to_numeric(ds['dropout_risk'], errors='coerce')

x=ds['dropout_risk'].mean()
print(x)
ds['dropout_risk'].fillna(x,inplace=True)
```

```python
#saving cleaned dataset
ds.to_csv('dropout_risk_cleaned.csv')
```
